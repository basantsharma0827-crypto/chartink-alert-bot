```python
# ==========================================================
# AlphaMind
# Cell 11 : Universal History Downloader
# ==========================================================

from datetime import datetime, timedelta
import pandas as pd

def download_history(symbol,
                     resolution="D",
                     days=365):

    today = datetime.today()
    start = today - timedelta(days=days)

    data = {
        "symbol": symbol,
        "resolution": resolution,
        "date_format": "1",
        "range_from": start.strftime("%Y-%m-%d"),
        "range_to": today.strftime("%Y-%m-%d"),
        "cont_flag": "1"
    }

    response = fyers.history(data)

    if response["s"] != "ok":
        print(response)
        return None

    df = pd.DataFrame(
        response["candles"],
        columns=[
            "timestamp",
            "open",
            "high",
            "low",
            "close",
            "volume"
        ]
    )

    df["datetime"] = pd.to_datetime(df["timestamp"], unit="s")

    df = df[
        [
            "datetime",
            "open",
            "high",
            "low",
            "close",
            "volume"
        ]
    ]

    return df
```

```python
# ==========================================================
# AlphaMind
# Cell 17 : Check if Data Already Exists
# ==========================================================

def is_downloaded(symbol, timeframe):

    query = f"""
        SELECT COUNT(*)
        FROM metadata
        WHERE symbol='{symbol}'
        AND timeframe='{timeframe}'
    """

    count = con.execute(query).fetchone()[0]

    return count > 0

print("✅ is_downloaded() function created.")
```

```python
# ==========================================================
# AlphaMind
# Cell 21 : Save Any Stock to Database
# ==========================================================

def save_to_database(symbol, timeframe, df):

    # Check if already downloaded
    if is_downloaded(symbol, timeframe):
        print(f"⏩ {symbol} ({timeframe}) already exists. Skipping...")
        return

    temp = df.copy()
    temp["symbol"] = symbol

    con.register("temp_df", temp)

    con.execute("""
        INSERT INTO historical_data
        SELECT
            symbol,
            datetime,
            open,
            high,
            low,
            close,
            volume
        FROM temp_df
    """)

    con.execute(f"""
        INSERT OR REPLACE INTO metadata
        VALUES(
            '{symbol}',
            '{timeframe}',
            '{df['datetime'].min().date()}',
            '{df['datetime'].max().date()}',
            CURRENT_TIMESTAMP,
            {len(df)},
            'COMPLETE'
        )
    """)

    print(f"✅ {symbol} saved successfully.")
```

```python
# ==========================================================
# AlphaMind
# Cell 23 : Universal Downloader
# ==========================================================

def download_and_save(symbol,
                      timeframe="D",
                      days=365):

    # Already downloaded?
    if is_downloaded(symbol, timeframe):
        print(f"⏩ {symbol} already exists.")
        return

    print(f"⬇ Downloading {symbol}...")

    df = download_history(
        symbol=symbol,
        resolution=timeframe,
        days=days
    )

    if df is None:
        print(f"❌ Failed : {symbol}")
        return

    save_to_database(
        symbol,
        timeframe,
        df
    )

    print(f"✅ Finished : {symbol}")
```

```python
# ==========================================================
# AlphaMind
# Cell 28 : Market Engine
# ==========================================================

from tqdm import tqdm
import time

def market_engine(symbols,
                  timeframe="D",
                  days=365,
                  sleep_time=0.25):

    success = 0
    failed = 0
    skipped = 0

    for symbol in tqdm(symbols):

        try:

            if is_downloaded(symbol, timeframe):

                skipped += 1
                continue

            df = download_history(
                symbol=symbol,
                resolution=timeframe,
                days=days
            )

            if df is None:

                failed += 1
                continue

            save_to_database(
                symbol,
                timeframe,
                df
            )

            success += 1

            time.sleep(sleep_time)

        except Exception as e:

            print(symbol, e)

            failed += 1

    print("\n==============================")
    print("Download Finished")
    print("==============================")
    print("Success :", success)
    print("Skipped :", skipped)
    print("Failed  :", failed)
```

```python
# ==========================================================
# AlphaMind
# Cell 32 : Master Downloader Engine
# ==========================================================

from tqdm import tqdm
from datetime import datetime
import time

def download_market(timeframe="D", days=365):

    symbols = con.execute("""
        SELECT symbol
        FROM symbols
        WHERE status='ACTIVE'
    """).fetchall()

    symbols = [i[0] for i in symbols]

    total = len(symbols)

    print(f"Total Symbols : {total}")

    success = 0
    failed = 0
    skipped = 0

    for symbol in tqdm(symbols):

        try:

            if is_downloaded(symbol, timeframe):

                skipped += 1
                continue

            df = download_history(
                symbol=symbol,
                resolution=timeframe,
                days=days
            )

            if df is None:

                failed += 1

                con.execute(f"""

                INSERT INTO failed_downloads
                VALUES(

                '{symbol}',

                '{timeframe}',

                'Download Failed',

                1,

                CURRENT_TIMESTAMP

                )

                """)

                continue

            save_to_database(
                symbol,
                timeframe,
                df
            )

            con.execute(f"""

            INSERT INTO download_log
            VALUES(

            NULL,

            '{symbol}',

            '{timeframe}',

            'SUCCESS',

            {len(df)},

            CURRENT_TIMESTAMP,

            CURRENT_TIMESTAMP,

            'OK'

            )

            """)

            success += 1

            time.sleep(0.25)

        except Exception as e:

            failed += 1

            print(symbol)

            print(e)

    print("\n==========================")
    print("DOWNLOAD COMPLETED")
    print("==========================")

    print("Success :", success)
    print("Skipped :", skipped)
    print("Failed :", failed)
```

```python
# ==========================================================
# AlphaMind
# Cell 37 : Feature Engine V1
# ==========================================================

import pandas as pd
import numpy as np

def feature_engine(df):

    df = df.copy()

    # Candle Features
    df["body"] = df["close"] - df["open"]

    df["body_pct"] = (
        abs(df["close"] - df["open"])
        /
        (df["high"] - df["low"] + 0.000001)
    )

    df["upper_wick"] = df["high"] - df[["open","close"]].max(axis=1)

    df["lower_wick"] = df[["open","close"]].min(axis=1) - df["low"]

    df["range"] = df["high"] - df["low"]

    # Price Change

    df["return_1"] = df["close"].pct_change()

    df["return_5"] = df["close"].pct_change(5)

    df["return_10"] = df["close"].pct_change(10)

    # Volume

    df["vol_avg20"] = df["volume"].rolling(20).mean()

    df["rel_volume"] = df["volume"] / df["vol_avg20"]

    return df

print("✅ Feature Engine Ready")
```

```python
from datetime import datetime, timedelta
import pandas as pd

def download_stock(symbol,
                   resolution="D",
                   start_date="2015-01-01",
                   end_date=None):

    if end_date is None:
        end_date = TODAY.strftime("%Y-%m-%d")

    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")

    all_data = []

    while start <= end:

        chunk_end = min(start + timedelta(days=365), end)

        data = {
            "symbol": symbol,
            "resolution": resolution,
            "date_format": "1",
            "range_from": start.strftime("%Y-%m-%d"),
            "range_to": chunk_end.strftime("%Y-%m-%d"),
            "cont_flag": "1"
        }

        print(f"Downloading {data['range_from']} → {data['range_to']}")

        response = fyers.history(data)

        if response["s"] != "ok":
            print(response)
            break

        all_data.extend(response["candles"])

        start = chunk_end + timedelta(days=1)

    if len(all_data) == 0:
        print("❌ No Data")
        return None

    df = pd.DataFrame(
        all_data,
        columns=[
            "timestamp",
            "open",
            "high",
            "low",
            "close",
            "volume"
        ]
    )

    df["datetime"] = pd.to_datetime(df["timestamp"], unit="s")
    df["symbol"] = symbol

    df = df[
        [
            "symbol",
            "datetime",
            "open",
            "high",
            "low",
            "close",
            "volume"
        ]
    ]

    df = df.drop_duplicates()

    print(f"✅ Total Candles : {len(df)}")

    return df
```

```python
# ==========================================================
# AlphaMind
# Cell 78 : Smart Incremental Downloader
# ==========================================================

from datetime import timedelta

def update_stock(symbol):

    last_date = get_last_date(symbol)

    if last_date is None:

        start_date = "2015-01-01"

    else:

        start_date = (last_date + timedelta(days=1)).strftime("%Y-%m-%d")

    end_date = TODAY.strftime("%Y-%m-%d")

    if start_date > end_date:

        print(f"✅ {symbol} already up to date.")

        return None

    print(f"Downloading {symbol}")
    print(f"From : {start_date}")
    print(f"To   : {end_date}")

    df = download_stock(
        symbol,
        start_date=start_date,
        end_date=end_date
    )

    if df is None:

        return None

    con.register("temp_df", df)

    con.execute("""
        INSERT INTO historical_data
        SELECT *
        FROM temp_df
    """)

    print(f"✅ Added {len(df)} new candles")

    return df
```

```python

# ==========================================================
# AlphaMind
# Cell 88 : Feature Table
# ==========================================================

con.execute("""

CREATE TABLE IF NOT EXISTS feature_data(

symbol VARCHAR,
datetime TIMESTAMP,

close DOUBLE,

ema5 DOUBLE,
ema10 DOUBLE,
ema20 DOUBLE,
ema50 DOUBLE,
ema200 DOUBLE,

rsi14 DOUBLE,

volume BIGINT,

PRIMARY KEY(symbol,datetime)

)

""")

print("✅ Feature Table Ready")
```

```python
import pandas as pd

def feature_engine(df):

    data = df.copy()

    data = data.sort_values("datetime")

    data["ema5"] = data["close"].ewm(span=5).mean()

    data["ema10"] = data["close"].ewm(span=10).mean()

    data["ema20"] = data["close"].ewm(span=20).mean()

    data["ema50"] = data["close"].ewm(span=50).mean()

    data["ema200"] = data["close"].ewm(span=200).mean()

    delta = data["close"].diff()

    gain = delta.clip(lower=0)

    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(14).mean()

    avg_loss = loss.rolling(14).mean()

    rs = avg_gain / avg_loss

    data["rsi14"] = 100 - (100 / (1 + rs))

    return data
```

```python
# ==========================================================
# AlphaMind
# Cell 92 : Create AI Targets
# ==========================================================

def create_targets(df):

    data = df.copy()

    # Next day's close
    data["next_close"] = data["close"].shift(-1)

    # 5-day future close
    data["future_5"] = data["close"].shift(-5)

    # 10-day future close
    data["future_10"] = data["close"].shift(-10)

    # Returns
    data["return_1"] = ((data["next_close"]-data["close"])/data["close"])*100

    data["return_5"] = ((data["future_5"]-data["close"])/data["close"])*100

    data["return_10"] = ((data["future_10"]-data["close"])/data["close"])*100

    return data
```

```python
def backtest_ema_cross(df):

    trades=[]

    position=False

    entry_price=0
    entry_date=None

    for i in range(1,len(df)-5):

        if not position:

            if df.ema5.iloc[i]>df.ema20.iloc[i] and df.ema5.iloc[i-1]<=df.ema20.iloc[i-1]:

                position=True
                entry_price=df.close.iloc[i]
                entry_date=df.datetime.iloc[i]

        else:

            exit_price=df.close.iloc[i+5]
            exit_date=df.datetime.iloc[i+5]

            ret=((exit_price-entry_price)/entry_price)*100

            trades.append([
                "EMA5_EMA20",
                df.symbol.iloc[i],
                entry_date,
                entry_price,
                exit_date,
                exit_price,
                ret,
                5,
                ret>0
            ])

            position=False

    return pd.DataFrame(trades,columns=[
        "strategy_name",
        "symbol",
        "entry_date",
        "entry_price",
        "exit_date",
        "exit_price",
        "return_pct",
        "holding_days",
        "win"
    ])
```

```python
import pandas as pd
from datetime import datetime

def analyze_results(result):

    total = len(result)

    wins = len(result[result["return_pct"] > 0])

    losses = total - wins

    win_rate = round((wins / total) * 100, 2) if total else 0

    avg_return = round(result["return_pct"].mean(), 2) if total else 0

    best_return = round(result["return_pct"].max(), 2) if total else 0

    worst_return = round(result["return_pct"].min(), 2) if total else 0

    gross_profit = result[result["return_pct"] > 0]["return_pct"].sum()

    gross_loss = abs(result[result["return_pct"] < 0]["return_pct"].sum())

    profit_factor = round(gross_profit / gross_loss, 2) if gross_loss > 0 else 999

    summary = pd.DataFrame([{
        "strategy_name": "EMA5_EMA20",
        "total_trades": total,
        "wins": wins,
        "losses": losses,
        "win_rate": win_rate,
        "avg_return": avg_return,
        "best_return": best_return,
        "worst_return": worst_return,
        "profit_factor": profit_factor,
        "created_at": datetime.now()
    }])

    return summary
```

```python
# ==========================================================
# AlphaMind
# Cell 123 : Generate Features Automatically
# ==========================================================

import pandas as pd
import numpy as np

def generate_features(df):

    df = df.copy()

    df = df.sort_values("datetime")

    # EMA
    df["ema5"] = df["close"].ewm(span=5).mean()
    df["ema10"] = df["close"].ewm(span=10).mean()
    df["ema20"] = df["close"].ewm(span=20).mean()
    df["ema50"] = df["close"].ewm(span=50).mean()
    df["ema100"] = df["close"].ewm(span=100).mean()
    df["ema200"] = df["close"].ewm(span=200).mean()

    # SMA
    df["sma20"] = df["close"].rolling(20).mean()
    df["sma50"] = df["close"].rolling(50).mean()

    # Volume MA
    df["volume_ma20"] = df["volume"].rolling(20).mean()

    # Relative Volume
    df["relative_volume"] = df["volume"] / df["volume_ma20"]

    # Daily Return
    df["daily_return"] = df["close"].pct_change()*100

    return df
```

```python
# ==========================================================
# AlphaMind
# Cell 134 : Latest Candle Scanner
# ==========================================================

def latest_scan(query):

    sql = f"""

    SELECT *

    FROM (

        SELECT *,
               ROW_NUMBER() OVER(
                   PARTITION BY symbol
                   ORDER BY datetime DESC
               ) rn

        FROM feature_data

    )

    WHERE rn=1
    AND ({query})

    ORDER BY symbol

    """

    return con.execute(sql).fetchdf()
```

```python
import pandas as pd
import numpy as np

def generate_features(df):

    df = df.copy()

    df = df.sort_values("datetime")

    # ==========================
    # EMA
    # ==========================
    for p in [5,10,20,50,100,200]:
        df[f"ema{p}"] = df["close"].ewm(span=p, adjust=False).mean()

    # ==========================
    # SMA
    # ==========================
    df["sma20"] = df["close"].rolling(20).mean()
    df["sma50"] = df["close"].rolling(50).mean()

    # ==========================
    # Volume
    # ==========================
    df["volume_ma20"] = df["volume"].rolling(20).mean()
    df["relative_volume"] = df["volume"] / df["volume_ma20"]

    # ==========================
    # RSI 14
    # ==========================
    delta = df["close"].diff()

    gain = delta.where(delta > 0, 0)

    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(14).mean()

    avg_loss = loss.rolling(14).mean()

    rs = avg_gain / avg_loss

    df["rsi14"] = 100 - (100 / (1 + rs))

    return df
```
